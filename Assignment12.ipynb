```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-4, 4, 0.4)
Y = np.arange(-4, 4, 0.4)
X, Y = np.meshgrid(X, Y)
Z = np.sqrt(X**4 + Y**4)
Z = X**4 + Y**4
ax.plot_surface(X, Y, Z, rstride=5, cstride=5, cmap='rainbow')
plt.show()

x = np.linspace(-4,4,400)
y = x ** 4 + 4
plt.grid()
plt.plot(x,y)
plt.show()
```
